1) Basic array Operation

Conditional Aggregates

Calculate the mean mpg for cars grouped by number of cylinders using numpy technique

```python
mpg = mpg.apply(pd.to_numeric, errors='coerce')
```

```python
cylinder = mpg['cylinders'].to_numpy()
```

```python
unique = np.unique(cylinder)
```

```python
for cyl in unique:
    mean_mpg = mpg[cylinder == cyl].mean()
```

```python
mean_mpg
```
