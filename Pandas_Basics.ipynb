Finding Anomalies 

identify potential outliers in the mpg column using the Interquartile Range(IQR) method. Specifically:

 * Calculate the IQR for mpg.
 * Define outliers as values less than Q1 - 1.5 * IQR or greater than Q3 + 1.5 * IQR
 * Create a DataFrame of cars classified as outliers, display car_name,mpg,and model_year

```python
q1 = auto_numeric['mpg'].quantile(0.25)
q3 = auto_numeric['mpg'].quantile(0.75)

IQR = q3-q1
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
print("lower bound",lower_bound)
print("Upper Bound",upper_bound)
print("IQR",IQR)
```

```python
outlier = (auto_numeric['mpg'] < lower_bound) | (auto_numeric['mpg'] > upper_bound)
```

```python
outliers = auto_numeric[outlier]
```

```python
result = outliers[['car name', 'mpg', 'model year']]
```
