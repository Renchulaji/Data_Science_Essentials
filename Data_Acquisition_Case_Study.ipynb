## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
import pandas as pd
import os

home_dir = os.path.expanduser("~")

data_science_path = os.path.join(home_dir, "Desktop", "Data Science")

hour_csv_path = os.path.join(data_science_path, "hour.csv")

df = pd.read_csv(hour_csv_path)

windspeed = df['windspeed']

Q1 = windspeed.quantile(0.25)
Q3 = windspeed.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = windspeed[(windspeed < lower_bound) | (windspeed > upper_bound)]

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Number of Outliers: {len(outliers)}")
print("\nOutlier Values:")
print(outliers)
```

## Q.14 Find the correlation between numerical columns and discuss any interesting relationships.

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

home_dir = os.path.expanduser("~")

data_science_path = os.path.join(home_dir, "Desktop", "Data Science")

hour_csv_path = os.path.join(data_science_path, "hour.csv")

df = pd.read_csv(hour_csv_path)

correlation_matrix = df.select_dtypes(include=['number']).corr()

print("Correlation Matrix:\n")
print(correlation_matrix)

plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
import pandas as pd
import os

home_dir = os.path.expanduser("~")

data_science_path = os.path.join(home_dir, "Desktop", "Data Science")

hour_csv_path = os.path.join(data_science_path, "hour.csv")

df = pd.read_csv(hour_csv_path)

season_mapping = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}

season_rentals = df.groupby('season')['cnt'].sum().reset_index()

season_rentals['season'] = season_rentals['season'].map(season_mapping)

season_rentals = season_rentals.sort_values(by='cnt', ascending=False)

print("Total Rentals by Season:")
print(season_rentals)

most_rented_season = season_rentals.iloc[0]['season']
print(f"\nSeason with highest rentals: {most_rented_season}")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
import pandas as pd
import os

home_dir = os.path.expanduser("~")

data_science_path = os.path.join(home_dir, "Desktop", "Data Science")

hour_csv_path = os.path.join(data_science_path, "hour.csv")

df = pd.read_csv(hour_csv_path)

hourly_rentals = df.groupby('hr')['cnt'].sum().reset_index()

peak_hour_row = hourly_rentals.loc[hourly_rentals['cnt'].idxmax()]
peak_hour = int(peak_hour_row['hr'])
peak_count = int(peak_hour_row['cnt'])

print(f"Peak Hour: {peak_hour}:00")
print(f"Total Rentals at Peak Hour: {peak_count}")
```

## Q.18 In which day of a week bikes rents out most?

```python
import pandas as pd
import os

# Get your home directory
home_dir = os.path.expanduser("~")

# Path to Desktop/Data Science folder
data_science_path = os.path.join(home_dir, "Desktop", "Data Science")

# Full path to hour.csv
hour_csv_path = os.path.join(data_science_path, "hour.csv")

# Load dataset
df = pd.read_csv(hour_csv_path)

# Weekday mapping
weekday_mapping = {
    0: 'Sunday',
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday'
}

# Group by weekday and sum rentals
weekday_rentals = df.groupby('weekday')['cnt'].sum().reset_index()

# Map weekday codes to names
weekday_rentals['weekday'] = weekday_rentals['weekday'].map(weekday_mapping)

# Sort by rentals descending
weekday_rentals = weekday_rentals.sort_values(by='cnt', ascending=False)

# Print results
print("Total Rentals by Weekday:")
print(weekday_rentals)

# Print day with highest rentals
most_rented_day = weekday_rentals.iloc[0]['weekday']
print(f"\nDay with highest rentals: {most_rented_day}")
```

## Q.20 What is the maximum temperature observed in each of the seasons?
